```python
def get_true(line):
    l = np.array(line[['l1','l2','l3','l4','l5']]) - np.array(line[['w1','w2','w3','w4','w5']])
    w = np.array(line[['w1','w2','w3','w4','w5']]) - np.array(line[['l1','l2','l3','l4','l5']])
    l_c = sum(l>=2)
    w_c = sum(w>=2)

    return w_c==3 and l_c==line['wl']-3
```

```python
index_true = scores_e.apply(get_true, axis=1)
```

```python
# divide the rating to different levels
def level(rating):
    level_range = [0,500,1000,1300,1600,1800,2000,2200,2400,2550,2800]
    for i in range(10):
        if level_range[i]<=rating<level_range[i+1]:
            return i+1
```

```python
get_true(scores_e.loc[29])
```

```python
scores_e = scores_e[index_true]
```

```python
scores_e['Ll']=scores_e['Lr'].apply(level)
scores_e['Wl']=scores_e['Wr'].apply(level)
```

```python
# calculate the mean and std of w_L ratio of different levels. 
wlratio_mean = np.zeros((10,10))
wlratio_std = np.zeros((10,10))
def wlratio(scores_e, wlratio_mean, wlratio_std):
    for i in range(10):
        for j in range(10):
            wr = scores_e[(scores_e['Ll']==i+1) & (scores_e['Wl']==j+1)]['win_ratio']
            lr = 1-scores_e[(scores_e['Ll']==j+1) & (scores_e['Wl']==i+1)]['win_ratio']
            w_r_l=wr.append(lr)
            wlratio_mean[i,j]=np.mean(w_r_l)
            wlratio_std[i,j]=np.sqrt(np.var(w_r_l))
wlratio(scores_e, wlratio_mean, wlratio_std)        
    
```
